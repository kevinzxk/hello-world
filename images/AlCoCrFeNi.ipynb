```python
# Structural stability as a function of concentration (for a fixed volume):
bcc_mask = test.CM.loc[idx[:],idx['str']]=='bcc'
bcc_selected = test.CM.loc[idx[bcc_mask],idx[:]]
fcc_mask = test.CM.loc[idx[:],idx['str']]=='fcc'
fcc_selected = test.CM.loc[idx[fcc_mask],idx[:]]
# Isolate certain volume:
which_sws = 2.65
bcc_sws_mask = bcc_selected.loc[idx[:],idx['SWS']]==which_sws
fcc_sws_mask = fcc_selected.loc[idx[:],idx['SWS']]==which_sws
bcc_sws_selected = bcc_selected.loc[idx[bcc_sws_mask],idx[:]]
fcc_sws_selected = fcc_selected.loc[idx[fcc_sws_mask],idx[:]]
bcc_energies = pd.DataFrame(bcc_sws_selected.loc[idx[:],idx['CC',1,1]]).join(bcc_sws_selected[['Etot']])
bcc_energies.columns = ['CC','Etot(bcc)']
# For the joining to work correctly, we have to reset the indexing
bcc_energies = bcc_energies.reset_index(drop=True)
bcc_fcc_energies = bcc_energies.join(fcc_sws_selected[['Etot']].reset_index(drop=True))
# Rename columns
bcc_fcc_energies.columns = ['CC','Etot_bcc','Etot_fcc']
delta = bcc_fcc_energies.sub(bcc_fcc_energies['Etot_bcc'],axis=0)[['Etot_fcc']]
# Give the delta column a new name. Columns must have unique names, when they are joined.
delta.columns = ['E_fcc-E_bcc']
bcc_fcc_energies = bcc_fcc_energies.join(delta)
# Finally, create a plot for our efforts:
bcc_fcc_energies.plot(x='CC',y='E_fcc-E_bcc')
# We can see, that bcc becomes the ground state structure (E_fcc - E_bcc > 0)
# with high Al concentrations. => In HEAs Al is bcc stabiliser.
```
